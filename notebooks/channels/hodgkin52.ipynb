```python
%load_ext autoreload
%autoreload 2

from jax import config
config.update("jax_enable_x64", True)
config.update("jax_platform_name", "cpu")

import os
os.environ["XLA_PYTHON_CLIENT_MEM_FRACTION"]=".8"

from typing import Tuple, Dict

import jax.numpy as jnp
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

from jaxley_mech.channels.hodgkin52 import Na, K, Leak, Na8States, K5States

import jaxley as jx
from jaxley import Compartment
```

```python
# stimulus

# Duration and step size.
dt = 0.05  # ms

i_delay = 5.0  # ms
i_dur = 25.0  # ms
i_offset = .0
t_max = 50
current = jx.step_current(i_delay, i_dur, .002, dt, t_max, i_offset=i_offset)[:-1]

time_vec = jnp.arange(0, t_max, dt)
```

```python
# original HH model
cell_hh = Compartment()

na_hh = Na(name="Na")
k_hh = K(name="K")
leak = Leak()

cell_hh.insert(na_hh)
cell_hh.insert(k_hh)
cell_hh.insert(leak)

params = {
    "Na_gNa": 120e-3,  # S/cm^2
    "Na_eNa": 50,  # mV

    "K_gK": 36e-3,  # S/cm^2
    "K_eK": -77,  # mV

    "Leak_gLeak": 0.3e-3,  # S/cm^2
    "Leak_eLeak": -54.3,  # mV

    "length": 5,
    "radius": 2.5,
    "v": -65.0,
}

for channel_name, channel_param in params.items():
    cell_hh.set(channel_name, channel_param)

cell_hh.init_states()


state_names = ["Na_m", "Na_h", "K_n", "i_Na", "i_K", "i_Leak", "v"]
for state_name in state_names:
    cell_hh.record(state_name, verbose=False)

cell_hh.stimulate(current, verbose=False)
s_hh = jx.integrate(cell_hh, delta_t=dt)

# transform s_hh into dict
s_hh_dict = {}
for i, state_name in enumerate(state_names):
    s_hh_dict[state_name] = s_hh[i].T[:-1]

# markov formulation of HH model
cell_mm = Compartment()

na_mm = Na8States(name="Na", solver="newton")
k_mm = K5States(name="K", solver="newton")
leak = Leak()

cell_mm.insert(na_mm)
cell_mm.insert(k_mm)
cell_mm.insert(leak)

for channel_name, channel_param in params.items():
    cell_mm.set(channel_name, channel_param)
cell_mm.init_states()



state_names = ["Na_C3", "Na_C2", "Na_C1", "Na_I3", "Na_I2", "Na_I1", "Na_I", "Na_O", 
               "K_C4", "K_C3", "K_C2", "K_C1", "K_O", "i_Na", "i_K", "i_Leak", "v"]
for state_name in state_names:
    cell_mm.record(state_name, verbose=False)

# run the model
cell_mm.stimulate(current, verbose=False)
s_mm = jx.integrate(cell_mm, delta_t=dt)

# transform s_mm into dict
s_mm_dict = {}
for i, state_name in enumerate(state_names):
    s_mm_dict[state_name] = s_mm[i].T[:-1]

gNa_mm = na_mm.channel_params["Na_gNa"] * s_mm[0].T[:-1] 
gK_mm = k_mm.channel_params["K_gK"] * s_mm[1].T[:-1]
```

```python
# compare the two models
fig, ax = plt.subplot_mosaic(
    mosaic=[["I"], ["V"], ["i"], ["O"]], 
    figsize=(8, 12)
)

ax["I"].plot(time_vec, current, color='black')
ax["I"].set_ylabel("Current (nA)")

ax["V"].plot(time_vec, s_hh_dict["v"], label=r"$V^{HH}$")
ax["V"].plot(time_vec, s_mm_dict["v"], label=r"$V^{MM}$")
ax["V"].axhline(-65, color="black", lw=0.5, ls="--")
ax["V"].set_ylabel("V (mV)")
ax["V"].legend(frameon=False)

ax["i"].plot(time_vec, s_hh_dict["i_Na"], label=r"$I_{Na}^{HH}$", color="C0") 
ax["i"].plot(time_vec, s_hh_dict["i_K"], label=r"$I_{K}^{HH}$", color="C0", linestyle='--', ) 

ax["i"].plot(time_vec, s_mm_dict["i_Na"], label=r"$I_{Na}^{MM}$", color="C1" ) 
ax["i"].plot(time_vec, s_mm_dict["i_K"], color="C1", linestyle='--', label=r"$I_{K}^{MM}$") 
ax["i"].set_ylabel("Current (mA/cm$^2$)")
ax["i"].legend(frameon=False)

ax["O"].plot(time_vec, s_hh_dict["Na_m"] ** 3 * s_hh_dict["Na_h"], label=r"$m^4h$", color="C0")
ax["O"].plot(time_vec, s_hh_dict["K_n"] ** 4, label=r"$n^4$",  color="C0", linestyle='--')

ax["O"].plot(time_vec, s_mm_dict["Na_O"], label=r"$O_{Na}$",color="C1")
ax["O"].plot(time_vec, s_mm_dict["K_O"], label=r"$O_{K}$", color="C1", linestyle='--')
ax["O"].set_ylabel("Gate opening probability")
ax["O"].legend(frameon=False)

for a in ax.values():
    a.spines[["top", "right"]].set_visible(False)
    a.axvline(10, color="black", lw=0.5, ls="--")
    a.set_xlabel("Time (ms)")

fig.tight_layout()
fig.suptitle("Hodgkin-Huxley model vs. Markov States model", y=1.02)
```

## Conductance noise (FoxLu)

Simulate the Markov Na/K channels with conductance noise using the new SDE solver (EulerMaruyama) and compare against the deterministic Markov traces.

```python
# Markov SDE with explicit noise inputs
import jax

key = jax.random.PRNGKey(0)
key_na, key_k = jax.random.split(key)

noise_time = time_vec
current_noise = current[: len(noise_time)]

na_noise = Na8States(name="Na", solver="sde")
k_noise = K5States(name="K", solver="sde")
leak_noise = Leak()

# Use fewer channels so fluctuations are visible.
noise_params = params.copy()
noise_params.update({
    "Na_N_Na": 2_000.0,
    "K_N_K": 2_000.0,
})

v = noise_params["v"]
states_noise: Dict[str, jnp.ndarray] = {}
states_noise.update(na_noise.init_state(states_noise, v, noise_params, dt))
states_noise.update(k_noise.init_state(states_noise, v, noise_params, dt))

# Cylinder surface area (cm^2) including end caps; converts nA stimulus to mA/cm^2.
area_cm2 = 2 * jnp.pi * noise_params["radius"] * noise_params["length"] * 1e-8
area_cm2 += 2 * jnp.pi * (noise_params["radius"] ** 2) * 1e-8
C_m = noise_params.get("capacitance", 1.0)  # uF/cm^2

def renorm(vec):
    vec = jnp.clip(vec, 0.0)
    return vec / jnp.sum(vec)

xi_na = jax.random.normal(key_na, (len(noise_time), 8))
xi_k = jax.random.normal(key_k, (len(noise_time), 5))

v_trace = []
i_na_trace = []
i_k_trace = []
i_leak_trace = []
o_na_trace = []
o_k_trace = []

for t in range(len(noise_time)):
    na_step = na_noise.update_states(states_noise, dt, v, noise_params, xi=xi_na[t])
    na_vec = jnp.array(
        [
            na_step[f"Na_C3"],
            na_step[f"Na_C2"],
            na_step[f"Na_C1"],
            na_step[f"Na_O"],
            na_step[f"Na_I3"],
            na_step[f"Na_I2"],
            na_step[f"Na_I1"],
            na_step[f"Na_I"],
        ]
    )
    na_vec = renorm(na_vec)
    na_step = dict(
        zip(
            [
                "Na_C3",
                "Na_C2",
                "Na_C1",
                "Na_O",
                "Na_I3",
                "Na_I2",
                "Na_I1",
                "Na_I",
            ],
            na_vec,
        )
    )
    states_noise.update(na_step)

    k_step = k_noise.update_states(states_noise, dt, v, noise_params, xi=xi_k[t])
    k_vec = jnp.array([k_step[f"K_C4"], k_step[f"K_C3"], k_step[f"K_C2"], k_step[f"K_C1"], k_step[f"K_O"]])
    k_vec = renorm(k_vec)
    k_step = dict(zip(["K_C4", "K_C3", "K_C2", "K_C1", "K_O"], k_vec))
    states_noise.update(k_step)

    i_na = na_noise.compute_current(states_noise, v, noise_params)
    i_k = k_noise.compute_current(states_noise, v, noise_params)
    i_leak = leak_noise.compute_current(states_noise, v, noise_params)

    i_stim_density = current_noise[t] * 1e-6 / area_cm2
    dv = (i_stim_density - (i_na + i_k + i_leak)) / C_m
    v = v + dt * dv

    v_trace.append(v)
    i_na_trace.append(i_na)
    i_k_trace.append(i_k)
    i_leak_trace.append(i_leak)
    o_na_trace.append(states_noise["Na_O"])
    o_k_trace.append(states_noise["K_O"])

v_trace = jnp.stack(v_trace)
i_na_trace = jnp.stack(i_na_trace)
i_k_trace = jnp.stack(i_k_trace)
i_leak_trace = jnp.stack(i_leak_trace)
o_na_trace = jnp.stack(o_na_trace)
o_k_trace = jnp.stack(o_k_trace)

fig, ax = plt.subplot_mosaic(
    mosaic=[["I"], ["V"], ["i"], ["O"]],
    figsize=(8, 12),
)

ax["I"].plot(noise_time, current_noise, color="black")
ax["I"].set_ylabel("Current (nA)")

ax["V"].plot(time_vec, s_mm_dict["v"], label="Markov deterministic")
ax["V"].plot(noise_time, v_trace, label="Markov noisy (N=2e3)")
ax["V"].axhline(-65, color="black", lw=0.5, ls="--")
ax["V"].set_ylabel("V (mV)")
ax["V"].legend(frameon=False)

ax["i"].plot(noise_time, i_na_trace, label=r"$I_{Na}^{noise}$", color="C1")
ax["i"].plot(noise_time, i_k_trace, label=r"$I_{K}^{noise}$", color="C1", linestyle="--")
ax["i"].plot(time_vec, s_mm_dict["i_Na"], label=r"$I_{Na}^{det}$", color="C0")
ax["i"].plot(time_vec, s_mm_dict["i_K"], label=r"$I_{K}^{det}$", color="C0", linestyle="--")
ax["i"].set_ylabel("Current (mA/cm$^2$)")
ax["i"].legend(frameon=False)

ax["O"].plot(noise_time, o_na_trace, label=r"$O_{Na}^{noise}$", color="C1")
ax["O"].plot(noise_time, o_k_trace, label=r"$O_{K}^{noise}$", color="C1", linestyle="--")
ax["O"].plot(time_vec, s_mm_dict["Na_O"], label=r"$O_{Na}^{det}$", color="C0")
ax["O"].plot(time_vec, s_mm_dict["K_O"], label=r"$O_{K}^{det}$", color="C0", linestyle="--")
ax["O"].set_ylabel("Gate opening probability")
ax["O"].legend(frameon=False)

for a in ax.values():
    a.spines[["top", "right"]].set_visible(False)
    a.axvline(10, color="black", lw=0.5, ls="--")
    a.set_xlabel("Time (ms)")

fig.tight_layout()
fig.suptitle("Markov SDE conductance noise (Na/K)", y=1.02)
plt.show()
```
